```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

img = cv2.imread(r'C:\Users\X\Desktop\WhatsApp Image 2025-10-13 at 16.54.34_83d4c4af.jpg', cv2.IMREAD_GRAYSCALE)
mean = cv2.blur(img, (3,3)) #(3,3 is kernel size)
gaussian = cv2.GaussianBlur(img, (3,3), 0) #ksize and sigmaX(std in X direction)
median = cv2.medianBlur(img, ksize=3)
sobelx = cv2.Sobel(img, cv2.CV_32F, 1, 0, ksize=3)
sobely = cv2.Sobel(img, cv2.CV_32F, 0, 1, ksize=3)
sobel = cv2.magnitude(sobelx, sobely)

titles = ['Original', 'Mean', 'Gaussian', 'Median', 'Sobel']
images = [img, mean, gaussian, median, sobel]

plt.figure(figsize=(12,5))
for i in range (5) :
    plt.subplot(2,3,i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
    plt.show()
```
